## Importa√ß√£o do Pandas

```python
import pandas as pd
```

## Leitura do Dataset

```python
df = pd.read_csv('amazon_kdd_negative_subsampling.csv')
```

## Consulta do Cabe√ßalho

```python
df.head()
```

## Remo√ß√£o de Registros do Idioma Japon√™s

```python
df = df[(df["query_locale"] == 'us') & (df["product_locale"] == 'us')]
df
```

## Feature Engineering

```python
df = df[['product_id', 'product_title', 'product_description', 'query', 'esci_label', 'test', 'val']]
df
```

## Troca das Labels

```python
df['esci_label'] = df['esci_label'].replace({'exact': 2, 'substitute': 1, 'complement': 1, 'irrelevant': 0})
```

```python
df
```

## Substitui√ß√£o de float('NaN') por String Vazia

```python
df['product_description'] = df['product_description'].replace(float("NaN"), '')
```

```python
df
```

## Divis√£o do Dataset em Treino, Teste e Valida√ß√£o

```python
df_train = df[(df['test'] == 0) & (df['val'] == 0)]
df_val = df[(df['test'] == 0) & (df['val'] == 1)]
df_test = df[df['test'] == 1]
```

## Dataset de Treino

```python
df_train = df_train[['product_id', 'product_title', 'product_description', 'query', 'esci_label']]
df_train
```

## Dataset de Valida√ß√£o

```python
df_val = df_val[['product_id', 'product_title', 'product_description', 'query', 'esci_label']]
df_val
```

## Dataset de Teste

```python
df_test = df_test[['product_id', 'product_title', 'product_description', 'query', 'esci_label']]
df_test
```

## Exportando para arquivos CSV

```python
df_train.to_csv('df_train_amazon.csv', index=False)
df_val.to_csv('df_val_amazon.csv', index=False)
df_test.to_csv('df_test_amazon.csv', index=False)
```
