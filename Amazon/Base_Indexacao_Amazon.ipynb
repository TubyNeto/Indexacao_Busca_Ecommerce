## Importação do Pandas

```python
import pandas as pd
```

## Lendo o Dataset

```python
df_train = pd.read_csv('df_train_amazon.csv')
df_val = pd.read_csv('df_val_amazon.csv')
df_test = pd.read_csv('df_test_amazon.csv')
df = pd.concat([df_train, df_val, df_test])
```

## Cabeçalho do Dataset

```python
df.head()
```

## Remoção de Colunas

```python
df = df[['product_id', 'product_title', 'product_description']]
df
```

## Remoção de Linhas com Título faltante

```python
df.dropna(subset=['product_title'], inplace=True)
df
```

## Substituição de float('NaN') por String Vazia

```python
df['product_description'] = df['product_description'].replace(float("NaN"), '')
df
```

## Selecionando Produtos Únicos para Indexar

```python
df.drop_duplicates(inplace=True, subset='product_title')
df
```

## Salvando em um arquivo CSV

```python
df.to_csv('products_amazon.csv', index=False)
```
