```python
import pandas as pd
```

```python
results = pd.read_csv('result_search_amazon.csv')
```

```python
products = pd.read_csv('products_amazon.csv')
products
```

```python
results.drop(columns=results.columns[0], axis=1, inplace=True)
```

```python
df_test = pd.read_csv('df_test_amazon.csv')
df_test.head()
```

```python
results.replace(-1, '', inplace=True)
results
```

```python
relevance_matrix = pd.DataFrame()
```

```python
for column in results.columns:
    aux = []
    ground_truth = df_test[(df_test['query'] == column)]
    for item in results[column]: 
        if item in list(ground_truth['product_id']):
            aux.append(ground_truth[ground_truth['product_id'] == item]['esci_label'].iloc[0])
        else:
            aux.append(0.0)
    relevance_matrix[column] = aux
```

```python
relevance_matrix = relevance_matrix[:10]
```

## NDCG

```python
import numpy as np
```

```python
def ndcg(rel_true, rel_pred, p=10):
    
    """ Returns normalized Discounted Cumulative Gain
    Args:
        rel_true (1-D Array): relevance lists for particular user, (n_songs,)
        rel_pred (1-D Array): predicted relevance lists, (n_pred,)
        p (int): particular rank position
        form (string): two types of nDCG formula, 'linear' or 'exponential'
    Returns:
        ndcg (float): normalized discounted cumulative gain score [0, 1]
    """
    
    
    rel_true = np.sort(rel_true)[::-1]
    p = min( len(rel_true), min(len(rel_pred), p) )
    discount = 1 / (np.log2(np.arange(p) + 2))

   
    idcg = np.sum(rel_true[:p] * discount)
    dcg = np.sum(rel_pred[:p] * discount)
   
    return dcg / idcg

# NDCG-5,25
# Recall - 500
```

```python
ndcg_list = []
for column in relevance_matrix.columns:
    ground_truth = df_test[(df_test['query'] == column )].sort_values(by='esci_label', ascending=False)
    ndcg_list.append(ndcg(ground_truth['esci_label'],  list(relevance_matrix[column])))
```
