## Importação do Pandas

```python
import pandas as pd
```

## Lendo o Dataset

```python
df = pd.read_csv('wands_negative_subsampling.csv')
```

## Cabeçalho do Dataset

```python
df.head()
```

## Remoção de Colunas

```python
df = df[['product_id', 'query', 'product_name', 'product_description', 'label']]
df
```

## Remoção de Linhas com Título, Query ou Relevância faltantes

```python
df.dropna(subset=['product_name', 'query', 'label'], inplace=True)
df
```

## Substituição de float('NaN') por String Vazia

```python
df['product_description'] = df['product_description'].replace(float("NaN"), '')
df
```

## Troca das Labels

```python
df['label'].replace({0.5: 1, 1.0: 2, 0.0: 0}, inplace=True)
df
```

```python
df['label'] = df['label'].astype(int)
df
```

## Selecionando Produtos Únicos para Indexar

```python
df.drop_duplicates(inplace=True, subset='product_name')
df
```

## Seleção de Colunas para Indexação

```python
df = df[['product_id', 'product_name', 'product_description']]
df
```

## Salvando em um arquivo CSV

```python
df.to_csv('products_wands.csv', index=False)
```
