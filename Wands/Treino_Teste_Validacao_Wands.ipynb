## Importação do Pandas

```python
import pandas as pd
```

## Leitura do Dataset

```python
df = pd.read_csv('wands_negative_subsampling.csv')
```

## Consulta do Cabeçalho

```python
df.head()
```

## Feature Engineering

```python
df = df[['product_id', 'product_name', 'product_description', 'query', 'label', 'test', 'val']]
df
```

## Substituição de float('NaN') por String Vazia

```python
df['product_description'] = df['product_description'].replace(float("NaN"), '')
```

```python
df
```

## Divisão do Dataset em Treino, Teste e Validação

```python
df_train = df[(df['test'] == 0) & (df['val'] == 0)]
df_val = df[(df['test'] == 0) & (df['val'] == 1)]
df_test = df[df['test'] == 1]
```

## Dataset de Treino

```python
df_train = df_train[['product_id', 'product_name', 'product_description', 'query', 'label']]
df_train
```

## Dataset de Validação

```python
df_val = df_val[['product_id', 'product_name', 'product_description', 'query', 'label']]
df_val
```

## Dataset de Teste

```python
df_test = df_test[['product_id', 'product_name', 'product_description', 'query', 'label']]
df_test
```

## Exportando para arquivos CSV

```python
df_train.to_csv('df_train_wands.csv', index=False)
df_val.to_csv('df_val_wands.csv', index=False)
df_test.to_csv('df_test_wands.csv', index=False)
```
