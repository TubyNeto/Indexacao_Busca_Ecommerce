## Definição do Modelo

```python
from sentence_transformers import SentenceTransformer

model_id = 'tubyneto/crowdedflower-bert'
model = SentenceTransformer(model_id)
```

```python
import pandas as pd
df = pd.read_csv('result_search_crowdedflower.csv')
```

```python
df.drop(columns=['index'], axis=1, inplace=True)
df
```

```python
products = pd.read_csv('products_crowdedflower.csv')
products
```

```python
for column in df.columns:
    products_list = []
    for item in df[column]:
        if (item == -1):
            products_list.append('')
        else:
            product_title = products[products['_unit_id'] == item]['product_title'].iloc[0]
            products_list.append(product_title)
    df[column] = products_list
    
df  
```

```python
BertRanking = pd.DataFrame()
```

```python
from sentence_transformers import util

for i, column in enumerate(df.columns):

    query = [str(column)]
    products = list(df[column])
    products = list(filter(('').__ne__, products))
    sentences =  query + products
    embeddings = model.encode(sentences)

    list1 = products
    list2 = util.pytorch_cos_sim(embeddings[0], embeddings[1:]).tolist()[0]

    zipped_lists = zip(list2, list1)

    sorted_zipped_lists = sorted(zipped_lists, reverse=True)

    sorted_list1 = [element for _,element in sorted_zipped_lists]

    if (len(sorted_list1) >= 500):
        BertRanking[df.columns[i]] = sorted_list1
    else:
        sorted_list1 = sorted_list1 + ['']*(500 - len(sorted_list1))
        BertRanking[df.columns[i]] = sorted_list1
    
```

```python
BertRanking
```

```python
df_test = pd.read_csv('df_test_crowdedflower.csv')
```

```python
df_test
```

```python
relevance_matrix = pd.DataFrame()
```

```python
for column in df.columns:
    aux = []
    ground_truth = df_test[(df_test['query'] == column)]
    for item in BertRanking[column]: 
        if item in list(ground_truth['product_title']):
            aux.append(ground_truth[ground_truth['product_title'] == item]['relevance'].iloc[0])
        else:
            aux.append(0.0)
    relevance_matrix[column] = aux
```

```python
relevance_matrix
```

## NDCG

```python
import numpy as np
```

```python
def ndcg(rel_true, rel_pred, p=10):
    
    """ Returns normalized Discounted Cumulative Gain
    Args:
        rel_true (1-D Array): relevance lists for particular user, (n_songs,)
        rel_pred (1-D Array): predicted relevance lists, (n_pred,)
        p (int): particular rank position
        form (string): two types of nDCG formula, 'linear' or 'exponential'
    Returns:
        ndcg (float): normalized discounted cumulative gain score [0, 1]
    """
    
    
    rel_true = np.sort(rel_true)[::-1]
    p = min( len(rel_true), min(len(rel_pred), p) )
    discount = 1 / (np.log2(np.arange(p) + 2))

   
    idcg = np.sum(rel_true[:p] * discount)
    dcg = np.sum(rel_pred[:p] * discount)
   
    return dcg / idcg

# NDCG-5,25
# Recall - 500
```

```python
ndcg_list = []
for column in relevance_matrix.columns:
    ground_truth = df_test[(df_test['query'] == column )].sort_values(by='relevance', ascending=False)
    ndcg_list.append(ndcg(ground_truth['relevance'],  list(relevance_matrix[column])))
```

## NDCG em Ordem Decrescente 

```python
dict_ndcg = {}
for index, column in enumerate(relevance_matrix.columns):
    dict_ndcg[column] = ndcg_list[index]

df_ndcg = pd.DataFrame(dict_ndcg, index=['NDCG'])
df_ndcg = df_ndcg.transpose()
df_ndcg.sort_values(by='NDCG', ascending=False)
```

```python
df_ndcg['NDCG'].mean()
```
