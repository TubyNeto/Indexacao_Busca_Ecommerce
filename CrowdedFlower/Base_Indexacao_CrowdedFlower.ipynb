## Importação do Pandas

```python
import pandas as pd
```

## Lendo o Dataset

```python
df = pd.read_csv('crowdedflower.csv', encoding='ISO-8859-1')
```

## Cabeçalho do Dataset

```python
df.head()
```

## Remoção de Colunas

```python
df = df[['_unit_id', 'query', 'product_title', 'product_description', 'relevance']]
df
```

## Remoção de Linhas com Título, Query ou Relevância faltantes

```python
df.dropna(subset=['product_title', 'query', 'relevance'], inplace=True)
df
```

## Substituição de float('NaN') por String Vazia

```python
df['product_description'] = df['product_description'].replace(float("NaN"), '')
df
```

## Selecionando Produtos Únicos para Indexar

```python
df.drop_duplicates(inplace=True, subset='product_title')
df
```

## Seleção de Colunas para Indexação

```python
df = df[['_unit_id', 'product_title', 'product_description']]
df
```

## Salvando em um arquivo CSV

```python
df.to_csv('products_crowdedflower.csv', index=False)
```
