```python
df_test = pd.read_csv('df_test_crowdedflower2.csv')
df_test.head()
```

```python
relevance_matrix = pd.DataFrame()
```

```python
for column in results.columns:
    aux = []
    ground_truth = df_test[(df_test['query'] == column)]
    for item in results[column]: 
        if item in list(ground_truth['_unit_id']):
            aux.append(ground_truth[ground_truth['_unit_id'] == item]['relevance'].iloc[0])
        else:
            aux.append(0.0)
    relevance_matrix[column] = aux
```

## NDCG

```python
import numpy as np
```

```python
def ndcg(rel_true, rel_pred, p=10):
    
    """ Returns normalized Discounted Cumulative Gain
    Args:
        rel_true (1-D Array): relevance lists for particular user, (n_songs,)
        rel_pred (1-D Array): predicted relevance lists, (n_pred,)
        p (int): particular rank position
        form (string): two types of nDCG formula, 'linear' or 'exponential'
    Returns:
        ndcg (float): normalized discounted cumulative gain score [0, 1]
    """
    
    
    rel_true = np.sort(rel_true)[::-1]
    p = min( len(rel_true), min(len(rel_pred), p) )
    discount = 1 / (np.log2(np.arange(p) + 2))

   
    idcg = np.sum(rel_true[:p] * discount)
    dcg = np.sum(rel_pred[:p] * discount)
   
    return dcg / idcg

# NDCG-5,25
# Recall - 500
```

```python
ndcg_list = []
for column in relevance_matrix.columns:
    ground_truth = df_test[(df_test['query'] == column )].sort_values(by='relevance', ascending=False)
    ndcg_list.append(ndcg(ground_truth['relevance'],  list(relevance_matrix[column])))
```

## NDCG em Ordem Decrescente 

```python
dict_ndcg = {}
for index, column in enumerate(relevance_matrix.columns):
    dict_ndcg[column] = ndcg_list[index]

df_ndcg = pd.DataFrame(dict_ndcg, index=['NDCG'])
df_ndcg = df_ndcg.transpose()
df_ndcg.sort_values(by='NDCG', ascending=False)
```

## Média NDCG

```python
df_ndcg['NDCG'].mean()
```
